```python
#Loading companies and rounds dataset
import pandas as pd
rounds2_path = r'C:\Users\ganguso\Data Sc\Investment Assignment\rounds2.csv'
company_path = r'C:\Users\ganguso\Data Sc\Investment Assignment\companies.txt'
companies = pd.read_csv(company_path, sep='\t', encoding="ISO-8859-1")
rounds2 = pd.read_csv(rounds2_path, encoding = "ISO-8859-1")
```

```python
companies.head()
```

```python
rounds2.head()
```

```python
companies['permalink'] = companies['permalink'].str.lower()
rounds2['company_permalink'] = rounds2['company_permalink'].str.lower()
# printing the unique no. of companies from comapnies and rounds dataset
print(len(companies.permalink.unique()))
print(len(rounds2.company_permalink.unique()))
```

```python
#Merging companies to rounds data
master_frame = rounds2.merge(companies, left_on='company_permalink', right_on='permalink', how='inner')
master_frame.head()
```

```python
print(len(master_frame))
```

```python
#Dropping Null values
master_frame = master_frame.dropna()
master_frame.count()
```

```python
master_frame.count()
```

```python
master_frame.head()
```

```python
print("AVG of round type venture : "+ str(master_frame[master_frame['funding_round_type'].str.lower()=='venture']['raised_amount_usd'].mean()))
print("AVG of round type angel : "+ str(master_frame[master_frame['funding_round_type'].str.lower()=='angel']['raised_amount_usd'].mean()))
print("AVG of round type seed : "+ str(master_frame[master_frame['funding_round_type'].str.lower()=='seed']['raised_amount_usd'].mean()))
print("AVG of round type private equity : "+ str(master_frame[master_frame['funding_round_type'].str.lower()=='private_equity']['raised_amount_usd'].mean()))
```

```python
# Plotting the 1st plot : Fraction of total investments based on the mean value per FT type
from matplotlib import pyplot as plt
%matplotlib inline
list_fund = [master_frame[master_frame['funding_round_type'].str.lower()=='venture']['raised_amount_usd'].mean(),
                master_frame[master_frame['funding_round_type'].str.lower()=='angel']['raised_amount_usd'].mean(),
                master_frame[master_frame['funding_round_type'].str.lower()=='seed']['raised_amount_usd'].mean(),
                master_frame[master_frame['funding_round_type'].str.lower()=='private_equity']['raised_amount_usd'].mean()]
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.axis('equal')
f_type = ['Venture', 'Angel', 'Seed', 'Private Equity']
ax.pie(list_fund, labels = f_type, autopct='%1.2f%%')
plt.show()
```

```python
#Computing Top 9 countries for the chosen FT type
topx = master_frame[master_frame['funding_round_type'].str.lower()=='venture']
top9 = topx.groupby('country_code')[['raised_amount_usd']].sum().sort_values(by=['raised_amount_usd'], ascending=False)
top9 = top9[:9]
```

```python
print(top9)
```

```python
#Plotting top 9 countries based on the no. of investments type 
import numpy as np
plt.figure(1)
plt.title('Top 9 Countries')
plt.ylabel('Raised amount for venture type(Log Scale)')
plt.bar(list(top9.index),np.log(top9['raised_amount_usd']))
plt.show()
```

```python
master_frame = master_frame[master_frame['funding_round_type'].str.lower()=='venture'] 
master_frame['category_list'] = master_frame['category_list'].apply(lambda x : str(x).split("|")[0] if "|" in x else x)                                   
master_frame.head()
```

```python
#Loading the mapping file to extract one of the 8 main sectors for each of the primary sectors
sector_mapping_path = r'C:\Users\ganguso\Data Sc\Investment Assignment\mapping.csv'
sector_df = pd.read_csv(sector_mapping_path, encoding = "ISO-8859-1")
sector_df = sector_df.dropna()
sector_df.head()
```

```python
# Using pandas.melt() function to remove the one hot encoding in order to bring down the columns
sector_df = sector_df.melt(id_vars="category_list",var_name="main_sector")
sector_df = sector_df.loc[~(sector_df["value"]==0) ]
sector_df=sector_df.drop(columns="value")
```

```python
sector_df.head()
```

```python
#Merging main sectors
master_frame = master_frame.merge(sector_df, left_on='category_list', right_on='category_list', how='inner')
master_frame.head()
```

```python
#No. of investments and raised amount for Country 1 sector wise
D1 = master_frame[(master_frame['country_code']=='USA') & (master_frame['raised_amount_usd'] >= 5000000.00) & (master_frame['raised_amount_usd'] <= 15000000.00)]
print(D1.groupby('main_sector')['raised_amount_usd'].count().sort_values(ascending=False))
print(D1.groupby('main_sector')['raised_amount_usd'].count().sum())
print('*********************************************************')
print(D1.groupby('main_sector')['raised_amount_usd'].sum().sort_values(ascending=False))
print(D1.groupby('main_sector')['raised_amount_usd'].sum().sum())
```

```python
#No. of investments and raised amount for Country 2 sector wise
D2 = master_frame[(master_frame['country_code']=='GBR') & (master_frame['raised_amount_usd'] >= 5000000.00) & (master_frame['raised_amount_usd'] <= 15000000.00)]
print(D2.groupby('main_sector')['raised_amount_usd'].count().sort_values(ascending=False))
print(D2.groupby('main_sector')['raised_amount_usd'].count().sum())
print('*********************************************************')
print(D2.groupby('main_sector')['raised_amount_usd'].sum().sort_values(ascending=False))
print(D2.groupby('main_sector')['raised_amount_usd'].sum().sum())
```

```python
#No. of investments and raised amount for Country 3 sector wise
D3 = master_frame[(master_frame['country_code']=='IND') & (master_frame['raised_amount_usd'] >= 5000000.00) & (master_frame['raised_amount_usd'] <= 15000000.00)]
print(D3.groupby('main_sector')['raised_amount_usd'].count().sort_values(ascending=False))
print(D3.groupby('main_sector')['raised_amount_usd'].count().sum())
print('*********************************************************')
print(D3.groupby('main_sector')['raised_amount_usd'].sum().sort_values(ascending=False))
print(D3.groupby('main_sector')['raised_amount_usd'].sum().sum())
```

```python
D1['Count_of_investments_per_sector'] = D1.groupby('main_sector')['raised_amount_usd'].count()
D1['Sum_of_investments_per_sector'] = D1.groupby('main_sector')['raised_amount_usd'].sum()
D2['Count_of_investments_per_sector'] = D2.groupby('main_sector')['raised_amount_usd'].count()
D2['Sum_of_investments_per_sector'] = D2.groupby('main_sector')['raised_amount_usd'].sum()
D3['Count_of_investments_per_sector'] = D3.groupby('main_sector')['raised_amount_usd'].count()
D3['Sum_of_investments_per_sector'] = D3.groupby('main_sector')['raised_amount_usd'].sum()
```
